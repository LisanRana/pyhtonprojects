# how to install opencv properly

```python
import cv2 as cv
```

```python
img=cv.imread("groups.jpg")

```

```python
img
```

```python
img.shape
```

```python
#step3:convert the image to grayscale
gray_image=cv.cvtColor(img,cv.COLOR_BGR2GRAY)
gray_image
```

```python
gray_image.shape
```

```python
# step4:load the classifier
face_classifier =cv.CascadeClassifier(
     cv.data.haarcascades +"haarcascade_frontalface_default.xml")
```

```python
#perform the face detection
face = face_classifier.detectMultiScale(
    gray_image, scaleFactor=1.1, minNeighbors=5) 
```

```python
#displaying the image
for (x,y,w,h) in face:
        cv.rectangle(img,(x,y),( x + w, y + h),(0, 255, 0), 4)
```

```python
img_rgb=cv.cvtColor(img,cv.COLOR_BGR2RGB)
```

```python
import matplotlib.pyplot as plt
plt.figure(figsize=(20,10))
plt.imshow(img_rgb)
plt.axis('off')
```

# #Access the webcam


```python
video_capture=cv.VideoCapture(0)
```

```python
def detect_bounding_box(vid):
    gray_image=cv.cvtColor(vid,cv.COLOR_BGR2GRAY)
    faces=face_classifier.detectMultiScale(gray_image, 1.1,5,minSize=(40,40))
    for (x,y,w,h) in faces:
         cv.rectangle(vid, (x,y), ( x + w, y + h),(0, 255, 0), 4)
    return faces 
    
```

```python
while True:
    result,video_frame=video_capture.read()
    if result is False:
        break
        faces=detect_bound-box(
        video_frame
        )
        cv.imshow(
        "My Face Detection Project", video_frame
        ) #display the processed frame in a window named "My face detection project"
        if cv.waitkey(1) & 0xFF ==ord("q"):
            break
    video_capture.release()
    cv.destroyAllWindows()
```
